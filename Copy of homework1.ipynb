Problem 0. Write a function `DifferByOne()` which takes two numbers and returns `True` if they differ by exactly one, and `False` otherwise. You may accomplish this by using a conditional, but you don't have to. (Python's built-in `abs()` function may be helpful.)

Example input/output:
```
DifferByOne(2,3)
> True

DifferByOne(5,4)
>True

DifferByOne(7,10)
>False
```


````python
def DifferByOne(x,y):
  '''Function returns true if the difference between x and y is 1.'''
  return abs(y - x) == 1
````

````python
DifferByOne(20,21)
````

Problem 1. Write a function `Ascending()` that takes three numbers and returns `True` if they are in ascending order.
```
Ascending(3,6,10)
>True

Ascending(5,2,8)
>False
```


````python
def Ascending(x,y,z):
  '''function returns True if numbers are listed in ascending order'''
  numbers = (x,y,z)
  print(sorted(numbers))
  return sorted(numbers) == [x,y,z]

#Ascending(3,6,10)
#Ascending(5,2,8)
````

Problem 2. Write a function `Median()` which takes three numbers and returns their median (the number which is in the middle if they were put in order.)
```
Median(4,2,7)
>4

Median(3,3,9)
>3
```

````python
def Median(x,y,z):
    numbers = [x,y,z]
    numbers.sort()
    return numbers[1]

print(Median(4,2,7))
print(Median(10,9,9))
````

Problem 3. Write a function `Round()` that takes a number and rounds to the nearest integer.
